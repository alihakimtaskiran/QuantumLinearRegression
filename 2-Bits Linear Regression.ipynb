# 2-Bits of Linear Regression 
## Imports

```python
from qiskit import execute, Aer, QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_histogram, plot_state_qsphere, plot_state_city
import numpy as np
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:90% !important; }</style>"))
```

## Create Backends

```python
bv= Aer.get_backend("statevector_simulator")
bp= Aer.get_backend("qasm_simulator")
```

## Create Line Function

```python
a=QuantumRegister(2,"a")
x=QuantumRegister(2,"x")
y=QuantumRegister(4,"y")
c_mul=QuantumRegister(3,"c_mul")
c_add=QuantumRegister(2,"c_add")
label=QuantumRegister(4,"label")
loss=QuantumRegister(4,"loss")
result=ClassicalRegister(4,"result")
qc=QuantumCircuit(a,x,y,label,loss,c_mul,c_add,result)

for i in range(4):
    qc.h(i)

for i in range(8,12):
    qc.h(i)

qc.ccx(0,2,4)
qc.ccx(0,3,16)
qc.ccx(1,2,17)
qc.ccx(1,3,18)

qc.cx(16,19)
qc.cx(17,19)
qc.ccx(16,17,20)
qc.cx(19,5)
qc.cx(20,6)
qc.cx(18,6)
qc.ccx(18,20,7)

display(qc.draw("mpl"))
```
